## Listas

Listas em python representam uma seuqencia de valores em ordem, como por exemplo, representada entre clchetes e separdas por virgula:

> primes = [2, 3, 5, 7]

Pode-se colocar outros tipos dentro de listas:

> planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']

Podemos fazer lista de listas:

> hands = [
>     ['J', 'Q', 'K'],
>     ['2', '2', '2'],
>     ['6', 'A', 'K'], # (Comma after the last element is optional)
>]

Pode ser escrito também em uma linha mas dificulta a leitura:

> hands = [['J', 'Q', 'K'], ['2', '2', '2'], ['6', 'A', 'K']]

Em python a lista pode conter tipos de variáveis diferentes:

> my_favourite_things = [32, 'raindrops on roses', help]

## Indexação

Pode-se acessar nas listas um leemento iindividual, colocando seu indice/index entre colchetes:
You can access individual list elements with square brackets.

No python o index, começão do zero, sendo esse o primeiro elemento.

> planets[0]

Os elementos ao final de uma lista, podem ser acessador a partir do -1, e reduzindo -2, -3...

> planets[-1]

## Repartindo

Se quisermos pegar a lista dos 3 primeiros planetas, como poderiamos fazer essa divisão:

> planets[0:3]
> /['Mercury', 'Venus', 'Earth']

Usar o planets[0:3] é uma forma de cortar uma lista onde inclui o primeiro elemento 0 até o indice 3, este não estando incluso.

O índice incial e final são opcionais, se deixar tirar o inicial ele assume o valor de 0, podendo escrever da seguinte forma:

> planets[:3]

Se deixarmos sem o índice final ele assume o valor do comprimento da lista:

> planets[3:]
> //['Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']

Isto quer dizer, me dê todos os planetas a partir do indíce 3 para frente.

Pode-se utilizar também valores negativos para repartir listas:

Por exemplo para exibir todos os valores exceto o primeiro e o último:

>planets[1:-1]
>// ['Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus']

Ou os últimos 3 planetas:

> planets[-3:]

## Mudando listas

Listas são mut[aveis, significando que podem ser modificadas. Uma forma de modificar é associar a lista a um índice ou um pedaço de lista.

Por exemplo, se quisermos renomear o nome de Marte na lista de planetas:

> planets[3] = 'Malacandra'

Se quisermos encurtar o nome dos primeiros 3 planetas:

> planets[:3] = ['Mur', 'Vee', 'Ur']

Agora se quisermos voltar todos os nomes originais:

> planets[:4] = ['Mercury', 'Venus', 'Earth', 'Mars',]

## Funções de listas

Agora seguem algumas funções ou métodos atreladas as listas:

len = tamanho da lista;

> len(planets)

sorted, retorna a lista ordenada:

> sorted(planets)

sum, soma os valores

> sum(primes)

max e min, autoexplicativo:

> max(primes)


## Interlude: objects

Tudo em python é um objeto, isto é, carregam coisas ao redor dele. É possível acessar coisas usando uma sintax após o ponto depois do objeto. 

Por exemplo os números em Python carregam associado a eles uma variável imag, representando sua parte imaginária. 

>  x = 12

> print(x.imag)

> c = 12 + 3j

> print(c.imag)

Os objetos podem também carregar funções, uma função atrelada a um objeto é chamada de método. Algo não que não seja uma função, como o imag e atrelados a um objeto é chamado atributo.

Por exemplo, números tem um método chamado bit_length:

> x.bit_length()

## Métodos de lista

.append(), modifica a lista adicionando um item ao final dela:

> planets.append('Pluto')

O append, não retorna, isto pode ser visto na descrição do método help(append)

list.pop() remove e retorna o último elemento de uma lista:

> planets.pop()

## Procurando uma lista

list.index(), retorna o índice de uma lista.

> planets.index('Earth')

Se perguntar de um elemento que não esteja na lista, vai retornar um ValueError, dizendo que este elemento não pertence.

Para evitar esse erro, pode-se usar o operador in, para saber se um valor pertence ou não (True or False) em uma lista:

> "Earth" in planets

Para conhecer outros métodos basta usar o comando help(list)


## Tuples
Tuplas são a mesma coisa que as listas, mas diferem em dois pontos:

1: A sintaxe para criar lista ao inves de colchete utiliza parenteses:

> t = (1, 2, 3)

2: Não podem ser modificadas, são imutáveis:

> t[0] = 100

Retorna um TypeError

Tuplas são frequentemente usada para funções que retornam múltiplos valores.

Por exemplo, o método as_integer_ratio() de objetos reais, retornam um numerador e denominador na forma de tuplas:

> x = 0.125
> x.as_integer_ratio()

Esse retorno de valores múltiplos podem ser individualmente dados como a seguir:

> numerator, denominator = x.as_integer_ratio()
> print(numerator / denominator)

Finalmente alguns insight do clássico Stupid Python Trick™ para trocar duas variáveis:

> a = 1
> b = 0
> a,b = b,a
> print(a,b)

## Loops

O for é o exemplo básico de loops, que são estrutura de repetição:

> planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
> for planet in planets:
>    print(planet, end=' ')
